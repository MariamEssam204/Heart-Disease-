**Meta Data**

Age: age of the patient [years] <br>
Sex: sex of the patient [M: Male, F: Female <br> 0: Normal]

ChestPainType: chest pain type <br>
    [TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]<br>
    RestingBP: resting blood pressure [mm Hg]

 Cholesterol: serum cholesterol [mm/dl] <br>
    FastingBS: fasting blood sugar <br> [1: if FastingBS > 120 mg/dl, 0: otherwise] <br>

 RestingECG: resting electrocardiogram results <br> [Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria]<br>

 MaxHR: maximum heart rate achieved [Numeric value between 60 and 202] <br>
    ExerciseAngina: exercise-induced angina [Y: Yes, N: No] <br>
    Oldpeak: oldpeak = ST [Numeric value measured in depression] <br>

ST_Slope: the slope of the peak exercise ST segment [Up: upsloping, Flat: flat, Down: downsloping]<br>
    HeartDisease: output class [1: heart disease, 0: Normal] <br> 

```python
df_droped = df.drop(['FastingBS','Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'], axis =1)
```

```python
sns.heatmap(df_droped.corr(),cmap = 'YlGnBu')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

heart_disease_df = df[df['HeartDisease'] != 0]
heart_disease_df['Sex'].value_counts().plot.pie(ax=axs[0, 0], autopct='%1.1f%%')
axs[0, 0].set_title('Pie Chart: Distribution of Sex in Heart Disease Patients',fontsize=9)

# Subplot 2: Distribution of ChestPainType
heart_disease_df['ChestPainType'].value_counts().plot.pie(ax=axs[0, 1], autopct='%1.1f%%')
axs[0, 1].set_title('Pie Chart: Distribution of ChestPainType in Heart Disease Patients',fontsize=9)

# Subplot 3: Distribution of RestingECG
heart_disease_df['RestingECG'].value_counts().plot.pie(ax=axs[1, 0], autopct='%1.1f%%')
axs[1, 0].set_title('Pie Chart: Distribution of RestingECG in Heart Disease Patients',fontsize=9)

# Subplot 4: Distribution of ExerciseAngina
heart_disease_df['ExerciseAngina'].value_counts().plot.pie(ax=axs[1, 1], autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart: Distribution of ExerciseAngina in Heart Disease Patients',fontsize=9)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
